# Terms used for "Korea"

This notebook shows a summary of all the terms in iDigBio's country column that might possibly be interpreted as "korea".

```python
from pyspark.sql.functions import col
```

```python
df = sqlContext.read.parquet("/guoda/data/idigbio-20170826T023307.parquet")
```

```python
(df
 .filter(col("country").like("%korea%"))
 .groupby(col("country"))
 .count()
 .orderBy(col("count"), ascending=False)
 .toPandas()
 )
```
